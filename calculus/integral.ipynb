```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, integrate

x = symbols('x')

fx = x**2 + 1

area_0_1 = integrate(fx, (x, 0, 1))
area_1_2 = integrate(fx, (x, 1, 2))

print(f"fx: [{fx}], x : [0~1] 의 면적 구하기 : {area_0_1}")
print(f"fx: [{fx}], x : [1~2] 의 면적 구하기 : {area_1_2}")
```

```python
# 그래프 그리기
x_vals = np.linspace(-0.5, 2.5, 400)
y_vals = x_vals**2 + 1

fig, ax = plt.subplots()
ax.plot(x_vals, y_vals, label='$f(x) = x^2 + 1$', color='blue')
ax.fill_between(np.linspace(0, 1, 100), np.linspace(0, 1, 100)**2 + 1, alpha=0.3, color='red', label='Integral [0,1]')
ax.fill_between(np.linspace(1, 2, 100), np.linspace(1, 2, 100)**2 + 1, alpha=0.3, color='purple', label='Integral [1,2]')

# 그래프 설정
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
ax.set_title('Integral Visualization')
ax.legend()
plt.show()
```
